POPULARITY BASED

```python
ratings_with_name=ratings.merge(books, on ='ISBN')
```

```python
ratings_with_name
```

```python
num_rating_df=ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num-ratings'},inplace=True)
num_rating_df
```

```python
print(ratings_with_name['Book-Rating'].dtype)
print(ratings_with_name['Book-Rating'].isnull().sum())
```

```python
ratings_with_name['Book-Rating'] = ratings_with_name['Book-Rating'].fillna(0)

# Calculate mean ratings by grouping 'Book-Title'
avg_rating_df = ratings_with_name.groupby('Book-Title')['Book-Rating'].mean().reset_index()

# Rename the 'Book-Rating' column to 'avg-ratings'
avg_rating_df.rename(columns={'Book-Rating': 'avg-ratings'}, inplace=True)

# Print the resulting DataFrame
print(avg_rating_df)
```

```python
popular_df=num_rating_df.merge(avg_rating_df,on ='Book-Title')
popular_df
```

```python
popular_df=popular_df[popular_df['num-ratings']>=250].sort_values('avg-ratings',ascending=False).head(50)
```

```python
popular_df = popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num-ratings','avg-ratings']]
```

COLLABORATIVE

```python
x=ratings_with_name.groupby('User-ID').count()['Book-Rating']>200
padhe_likhe= x[x].index
```

```python
filtered_rating=ratings_with_name[ratings_with_name['User-ID'].isin(padhe_likhe)]
```

```python
filtered_rating
```

```python
y=filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books=y[y].index
```

```python
final_rating=filtered_rating[ratings_with_name['Book-Title'].isin(famous_books)]
```

```python
pt=final_rating.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score=cosine_similarity(pt)
```

```python
def recommend(book_name):
    index = np.where(pt.index == book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[index])), key=lambda x:x[1], reverse=True)[1:8]
    data=[]
    for i in similar_items:
        item=[]
        temp_df=books[books['Book-Title']==pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title']))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author']))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M']))

        data.append(item)
    return data
```

```python
similarity_score[1]
```

```python
recommend('1984')
```
